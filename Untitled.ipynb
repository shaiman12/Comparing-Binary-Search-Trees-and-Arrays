```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = open("BSTAve.txt","r")
f2 = open ("ArrayAve.txt","r")
nArr = []
n2Arr = []
countArr = []
count2Arr = []
for i in range(500):
    y = f.readline().split()
    y2 = f2.readline().split()
    nArr.append(int(y[0]))
    countArr.append(float(y[1]))
    
    n2Arr.append(int(y2[0]))
    count2Arr.append(float(y2[1]))
plt.plot(nArr,countArr, label = 'Binary Search Tree')
plt.plot(n2Arr,count2Arr, label = 'Array')
plt.xlabel("Size of data structure")
plt.ylabel("Number of comparison operations")
plt.title("Average Case Situation")
plt.legend()

plt.show













f.close()
```
